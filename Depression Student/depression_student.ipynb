```python
import pandas as pd

# Veri setini yükleme
df = pd.read_csv("dataset/Depression Student Dataset.csv")
```

```python
# İlk birkaç satıra göz atın
print(df.head())
```

```python
# Eksik değerleri kontrol etme
print(df.isnull().sum())
```

```python
# Temel istatistikler
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bir sütunun dağılımını çizme
sns.histplot(df["Age"], kde=True)
plt.show()
```

```python
# Kategorik bir değişkenin çubuk grafiğini oluşturma
sns.countplot(data=df, x="Dietary Habits")
plt.show()
```

```python
# Korelasyon matrisi

# Sadece sayısal sütunları seçme
numeric_df = df.select_dtypes(include=["number"])

# Korelasyon matrisi oluşturma
correlation_matrix = numeric_df.corr()


sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Korelasyon Matrisi")
plt.show()
```

```python
# İki değişken arasındaki ilişkiyi scatterplot ile inceleme
sns.scatterplot(data=df, x="Study Hours", y="Academic Pressure")
plt.show()
```

```python
# Görselleri kaydetme
plt.savefig("grafik.png")
```

```python
# İşlenmiş veriyi kaydetme
df.to_csv("işlenmiş_veri.csv", index=False)
```
