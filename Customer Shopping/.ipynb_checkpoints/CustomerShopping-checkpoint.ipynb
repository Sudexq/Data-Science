```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Veri setini yükleme
df = pd.read_csv("dataset/shopping_trends.csv")
```

```python
# Veri setine ilk bakış
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
# Veri seti hakkında bilgi
print(df.info())
```

```python
# Sayısal veriler için istatistiksel özet
print(df.describe())
```

```python
# Eşsiz değerlerin gözlemi
for column in df.columns:
    print(f"{column} eşsiz değerler: {df[column].unique()}")
```

```python
# Değerlerin dağılımı
print(df['Age'].value_counts())  # Örneğin, kategorik bir sütun
```

```python
# Eksik değer sorgusu
print(df.isnull().sum())
```

```python
# Eksik değer oranını görmek için
print((df.isnull().sum() / len(df)) * 100)
```

```python
category_sales = df.groupby("Category")["Purchase Amount (USD)"].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=category_sales, x="Category", y="Purchase Amount (USD)", palette="viridis")
plt.title("Kategori Bazında Satışlar")
plt.xlabel("Kategori")
plt.ylabel("Toplam Satış (USD)")
plt.show()
```

```python
location_sales = df.groupby("Location")["Purchase Amount (USD)"].mean().reset_index()

# Görselleştirme
plt.figure(figsize=(12, 6))
sns.barplot(data=location_sales, x="Location", y="Purchase Amount (USD)", palette="coolwarm")
plt.title("Lokasyon Bazında Ortalama Satış")
plt.xlabel("Lokasyon")
plt.ylabel("Ortalama Satış (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=df, x="Age", y="Purchase Amount (USD)", hue="Category")
plt.title("Yaş ve Satış İlişkisi")
plt.xlabel("Yaş")
plt.ylabel("Satış Miktarı (USD)")
plt.show()
```

```python
payment_counts = df["Payment Method"].value_counts()

# Pasta grafiği
plt.figure(figsize=(8, 8))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title("Ödeme Yöntemlerinin Dağılımı")
plt.show()
```

```python
sns.boxplot(data=df, x="Discount Applied", y="Purchase Amount (USD)")
plt.title("İndirimlerin Satış Miktarına Etkisi")
plt.xlabel("İndirim Uygulandı mı?")
plt.ylabel("Satış Miktarı (USD)")
plt.show()
```

```python
season_category_sales = df.groupby(["Season", "Category"])["Purchase Amount (USD)"].sum().unstack()

# Isı haritası
plt.figure(figsize=(10, 6))
sns.heatmap(season_category_sales, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Sezon ve Kategori Bazında Satışlar")
plt.xlabel("Kategori")
plt.ylabel("Sezon")
plt.show()
```
